```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import requests
import scipy
import PIL

#two different types of plots on the same axes
x =	np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1	= np.array([110, 180, 240, 99, 220])
y2	= np.array([120, 170, 230, 100,	210])
fig, ax	= plt.subplots(figsize=(7, 3))
ax.bar(x, y1, label = "person 1", color='indigo')
ax.plot(x, y2, label = "person 2", color='red')
```

```python
y2_sd = np.std(y2) #calculate standard deviation for y2 values
fig, ax	= plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color='red')	

ax.errorbar(x, #our x values
            y2, #our y values
            yerr = y2_sd, #our y error values
            fmt = 'none')
```

```python
y2_sd = np.std(y2) #calculate standard deviation for y2 values
fig, ax	= plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color='red')	

ax.errorbar(x, #our x values
            y2, #our y values
            yerr = y2_sd, #our y error values
            fmt = 'none',
            color = 'blue', #error bar color
            linewidth = 2, #error bar line width
            capthick = 5,#error bar cap size
            capsize = 5) #error bar cap size
            #errorevery = 2)	#error bar error values if you want	to skip some points
```

```python
from PIL	import Image #to open images
import	requests #to download images
from io	import BytesIO #to convert images to bytes

response = requests.get("https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png")
image_file	= BytesIO(response.content) #convert the image to bytes
image = Image.open(image_file) #open the image

fig, ax = plt.subplots(figsize = (7, 3))
ax.plot(x, y2, color='hotpink')

ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])

ax_image.imshow(image) #add the image to the plot
ax_image.axis('off') #remove the axis from the image
```

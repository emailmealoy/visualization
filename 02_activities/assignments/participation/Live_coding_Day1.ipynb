```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
```

```python
np.random.seed(2103)
x = np.arange(50)
y = np.random.randint(0, 100, 50)

fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(x, y) #scatter plot
```

```python
np.random.seed(2103)
x = np.arange(50)
y = np.random.randint(0, 100, 50)

fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(x, y) #bar plot
```

```python
np.random.seed(2103)
x = np.arange(50)
y = np.random.randint(0, 100, 50)

fig, ax = plt.subplots(figsize=(6, 4))
#histogram
ax.hist(y, bins=10, edgecolor='black')
```

```python
fig,		ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)

ax.set_title('Total Growth Over Time') #title
ax.set_ylabel('Total Growth') #ylabel
ax.set_xlabel('Years since start') #xlabel
```

```python
## Customizing the font properties of the title, xlabel, and ylabel
# Using fontdict to set properties like family, color, and size
font1 = {'family': 'sans-serif', 'color': 'red', 'size': 16}
font2	= {'family': 'serif', 'color': 'green', 'size': 14}
font3	= {'family': 'monospace', 'color': 'blue', 'size': 12}

fig,		ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)

ax.set_title('Total Growth Over Time', fontdict=font1, loc='left',	fontweight='bold')
ax.set_ylabel('Total Growth',	fontdict=font2)
ax.set_xlabel('Years since start' , fontdict=font3)
```

```python
#	Customizing the marker style and color in a scatter plot
fig	, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, 
           y,
           marker = '>',
           color = 'indigo')
```

```python
#	Customizing the marker style and color in a line plot
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot	(x,
         y,
         marker	= '*',
         color	= 'orange',
         linestyle = '-.',
         linewidth = 3)
```

```python
#	Customizing the marker style and color in a line plot with specific marker face and edge colors
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot	(x,
         y,
         marker	= '*',
         color	= '#537890',
         linestyle = '-.',
         linewidth = 3,
         markersize = 13,
         markerfacecolor = '#92F456',
         markeredgecolor = '#281450')
```

```python
#Grid customization
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot	(x,
         y,
         marker	= '*',
         color	= '#537890',
         linestyle = '-.',
         linewidth = 3,
         markersize = 13,
         markerfacecolor = '#92F456',
         markeredgecolor = '#281450')
ax.grid (axis='y',	
         linestyle='--', 
         color='#F97321', 
         linewidth=1.5)
```
